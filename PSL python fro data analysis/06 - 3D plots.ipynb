# 3D plotting
Matlplotlib offers the possibility to do fairly simple **3d plots** using some of its 2d plotting functions, such as **plot** or **scatter**. <br />
For this, you just need to initiate the figure and declare one of its axis as a **3d projection**.

```python
from mpl_toolkits import mplot3d
fig = plt.figure()
ax = plt.axes(projection='3d')
```

As you can see, Matpltolib initiated a **3d ax** using the **object oriented approach**. Now that it is initiated, we can plot data inside.

```python
# Generating data
import numpy as np

# Initializing a random number generator (RNG)
rng = np.random.RandomState(0)

x = rng.rand(30)
y = rng.rand(30)
z = rng.rand(30)

# Initializing figure
fig = plt.figure()
ax = plt.axes(projection='3d')

# Plotting data
ax.scatter(x, y, z)

plt.show()
```

## Plotting surfaces 
With the 3D plot functionality of Matplotlib, you can display more than surfaces in different fashions. <br />
One of them is **"plot_surface()"**. It allows to **display the surface separating each point in color**. The color of each tile will be determined proportionally to the z value using a **colormap** (default is Blue).

```python
# x and y axis 
x = np.linspace(-10, 10, 100) 
y = np.linspace(-10, 10, 100) 
   
X, Y = np.meshgrid(x, y) 
Z = np.cos(X) + np.sin(Y) 
  
fig = plt.figure() 
ax = plt.axes(projection ='3d') 
ax.plot_surface(X, Y, Z, cmap ='viridis') 
plt.show()
```

In this situation, **np.meshgrid** allows to create a grid (points on a 2d plane) using both x and y vectors. The returned values will be the **coordinates of each points** of this grid, **in both dimensions** of the 2d plane.

```python
x = np.linspace(-10, 10, 100) 
y = np.linspace(-10, 10, 100) 
   
X, Y = np.meshgrid(x, y) 
print(X.shape)
print(Y.shape)
```

Thus, we can use those returned values to make a computation dependent on the coordinates of the points.

```python
Z = np.cos(X) + np.sin(Y) 
print(Z.shape)
```

This will give us the z position of each point of the grid, wich is defined as a function of its x and y position.

An other way of representing surfaces it using the function **"plot_wireframe()"**. With this function, only a grid separating the points is represented. It may be useful in situations when you need to see through your surface.

```python
# x and y axis 
x = np.linspace(-1, 1, 10) 
y = np.linspace(-1, 1, 10) 
   
X, Y = np.meshgrid(x, y) 
Z = np.cos(X) + np.sin(Y) 
  
fig = plt.figure() 
ax = plt.axes(projection ='3d') 
ax.plot_wireframe(X, Y, Z) 

plt.show()
```
