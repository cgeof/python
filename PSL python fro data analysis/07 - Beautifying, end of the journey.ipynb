# Beautifying
Matlplotlib has a rich inventory of properties that you can set on an ax to make it **more appealing**, or just **more readable**. To illustrate this, we will use the case of the following graph:

```python
import numpy as np
x = np.linspace(0, 10, 100)
y = 10 * np.sin(x) + 20

fig = plt.figure()
plt.plot(x, y)
plt.show()
```

See, a plot by itself is very abstract. It needs some additional elements to be understandable.<br /> 
Let's start by the basics, the x and y axis.

## x and y axis (z works too)
### Labels
Obviously, the first elements that we need to make a figure readable are correct labelling of the axis. For this, you can use the functions **"xlabel()"** and **"ylabel()"**, followed by the label.

```python
x = np.linspace(0, 10, 100)
y = 10 * np.sin(x) + 20

fig = plt.figure()
plt.plot(x, y)
plt.xlabel("Time (days)")
plt.ylabel("Temperature (A.U.)")
plt.show()
```

Notice how less abstract the graph looks now. <br />
NB: This is the MATLAB style interface. In the case of the object oriented one, you will use **"set_xlabel()"** and **"set_ylabel()"** instead.

### Limits
Matpltolib automatically sets your **x and y limits** to fit the integrality of the data that you decided to plot. <br />However, there will be some cases to set the limits of your x and y axis yourself.<br />
For this, you can use the functions **"xlim()"** and **"ylim()"**, followed by a **array** containing two scalars, the **lower bound** that you desire for your axis, and the **upper bound**.

```python
x = np.linspace(0, 10, 100)
y = 10 * np.sin(x) + 20

fig = plt.figure()
plt.plot(x, y)
plt.xlim([3, 8])
plt.ylim([12.5, 25])
plt.show()
```

This synthax is for the MATLAB style interface. In the case of the object oriented, you will use **"set_xlim()"** and **"set_ylim()"** instead.<br />
NB: You can invert you upper and lower bounds, it will invert the axis.

## Title
The title of the figure can be set using the function **"title()"** for the MATLAB style interface, and **"set_title()"** for the object oriented interface. 

```python
x = np.linspace(0, 10, 100)
y = 10 * np.sin(x) + 20

fig = plt.figure()
plt.plot(x, y)
plt.title("Timecourse of temperature")
plt.show()
```

## Lines
You can add a vertical or an horizontal line to your graphs. For this, you can use respectively the functions **"axvline()"** and **"axhline()"**, with the desired x or y value as a parameter.

```python
x = np.linspace(0, 10, 100)
y = 10 * np.sin(x) + 20

fig = plt.figure()
plt.plot(x, y)
plt.axhline(y=np.average(y), color="red")
plt.show()
```

NB: All the usual line customization parameters can be applied here.

# Subplots
Generating individual graphs can be tedious, specially when you are plotting graphs that are related and that you will want to assemble lately.<br />
Fortunately, Matpltolib has a very convenient feature called **subplots**.<br />
This allows to **split your figure in a Grid** with a given amount of **rows and columns**, where each cell will be able to contain a single graph.
You can initalize suplot axes using the function **"subplots()"**, with the desired number of rows and cols.

```python
x = np.linspace(0, 10, 100)
y = 10 * np.sin(x) + 20

f, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(x, y)
ax2.scatter(x, y)
plt.show()
```

An other way of intializing a subplot is to first initialize a figure, and then call the function **"subplot()"** to atribute a subplot to an ax.<br /> 
You will give as an argument the **number of rows** that you want, the **number of columns** and then the indice of the subplot in this Grid that you want the ax to refer to.<br /> 
This sequential initialization of subplots allows to have a better tweaking of the subplots.

```python
x = np.linspace(0, 10, 100)
y = 10 * np.sin(x) + 20

fig = plt.figure()
ax1 = plt.subplot(1,2,1)
ax2 = plt.subplot(1,2,2, sharex=ax1, sharey=ax1)
ax1.plot(x, y)
ax2.scatter(x, -1* y)
plt.show()
```

# Saving the figure
And then, to finish the journey of your figure, you can **save** it. <br /> 
Matplotlib allows you to save your figure in a lot of **different formats**, including **jpg, png or pdf**. <br /> 
For this, you can use the function **"savefig()"**, followed by the **name of the file** that you want to create (with the correct file extension at the end), and the **format** as a string.

```python
import numpy as np
from mpl_toolkits import mplot3d

x = np.linspace(-1, 1, 10) 
y = np.linspace(-1, 1, 10) 
   
X, Y = np.meshgrid(x, y) 
Z = np.cos(X) + np.sin(Y) 
  
fig = plt.figure() 
ax = plt.axes(projection ='3d') 
ax.plot_surface(X, Y, Z, cmap="viridis") 

plt.savefig("My_first_figure.pdf", format="pdf")
plt.show()
```
