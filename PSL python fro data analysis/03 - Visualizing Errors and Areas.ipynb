# Error plots
## Error bars
Sometimes, you need to **represent data in an aggregated way**, for example the average value. However, you know that this aggregation is not completely faithful to the reality. Thus, you will have an **error** associated to this aggregation.<br />
Typical examples of aggregation/error couples are average +/- SD, average +/- SEM, median +/- MAD. Usually, this error is represented by **error bars** surrounding data points.<br />
With Matplotlib, you can plot error bars using the **"errorbar()"** function. The error in y axis will be supplied with the **"yerr"** parameter.

```python
# Generating data
import numpy as np
x = np.linspace(0, 10, 25)
error = 0.3
y = np.cos(x) 

# Adding a random error
y += error * np.random.randn(25)

# Initializing figure
fig = plt.figure()

# Plotting
plt.errorbar(x, y, yerr=error)

#Showing
plt.show()
```

We introduced the yerr parameter to specify the error in y axis. Additionally, the parameter **"xerr"** can specify the **error on the x axis**.

```python
import numpy as np
x = np.linspace(0, 10, 25)
x_error = 0.2
y_error = 0.5
# TODO: Add a random error (from 0 to x_error) to x
x += x_error * np.random.RandomState(0).randn(25)

y = np.cos(x)
# TODO: Add a random error (from 0 to y_error) to y
y += y_error * np.random.RandomState(0).randn(25)

fig = plt.figure()
# TODO: Plot x and y, considering the error in x and y
plt.errorbar(x,y,xerr=x_error, yerr=y_err)
plt.show()
```

Until now we treated the error as a constant. However, error will vary depending on the data point. <br />
You can take this into account by specifying an **iterable** (list, array, tupple) having the correct number of values (number of data points) as **xerr and yerr**.

```python
x = np.arange(0, 12)
y_base = np.array([0, 2, 1, -5, -20, 5, 1, 2, 3, 5, 2, 1])
y = []

# Creates a list of 10 repetitions of y_base, with randomness added
for i in range(0, 10):
    y.append(y_base + np.random.randn(12))
print(y[0:2])
    

fig = plt.figure()    
# TODO: represent the mean and std of each column of y

plt.errorbar(x,np.average(y,axis=0), yerr=np.std(y,axis=0))
plt.show()
```

## Customization
The most creative part of making figures is to play with all the parameters that allow to customize the graph.
Let's see some of them.

### Format
You can drastically change the appearance of the errorbar plot using the **"fmt"** parameter, then specifying the format as a string.

```python
x = np.linspace(0, 10, 25)
error = 0.3
y = np.cos(x) 
y += error * np.random.randn(25)

fig = plt.figure()
# TODO: Create an errorbar plot represeting the error in y, using ".k" as a format
plt.errorbar(x,y,yerr=y_error, fmt=".k")
plt.show()
```

### Color
You can change the color of the line (and eventually marker) using the **"color"** argument, then specifying the wanted color as a string.

```python
x = np.linspace(0, 10, 25)
error = 0.3
y = np.cos(x) 
y += error * np.random.randn(25)

fig = plt.figure()
plt.errorbar(x, y, yerr=error, fmt="o", color="red")
plt.show()
```

### Alpha
Transparency can be added to the line using the **"alpha"** argument, then specifying the wanted transparency as a float between 0 and 1.
The default is 1 (opaque).

```python
x = np.linspace(10, 20, 25)
error = 0.5
y = np.cos(x) 
y += error * np.random.randn(25)

fig = plt.figure()
plt.errorbar(x, y, yerr=error, fmt="o", color="darkviolet", alpha=0.3)
plt.show()
```

### Error bar width and Capsize
You can adjust the width of the error bars using the **"elinewidth"** parameter (default 1). You can also change the size of the error bars caps using the **"capsize"** parameter (default 0, depends on style)

```python
x = np.linspace(30, 40, 25)
error = 0.2
y = np.cos(x) 
y += error * np.random.randn(25)

fig = plt.figure()
# TODO: Create an errorbar plot represeting the error in y, with custom elinewidth and capsize
plt.errorbar(x,y,yerr=y_error,elinewidth=2, capsize=10)
plt.show()
```

## Error areas
An other way of representing errors in a line plot is using **areas surrounding the line**. You can represent areas surrounding a line plot using the **"fill_between()"** function.<br />
This function takes as parameters **"x"**(from the central line), the **lower bound** and **upper bound**.<br />
Line parameters such as color and alpha can be applied to the fill_between.

```python
# Generating data
import numpy as np
x = np.linspace(0, 20, 1000)
error = 0.3 
y = np.cos(x) 
lower_bound = y - error
upper_bound = y + error


# Initializing figure
fig = plt.figure()

# Plotting the line first
plt.plot(x, y)

# TODO: Add the error using fill_between, with "lightgray" as color
plt.fill_between(x, lower_bound, upper_bound, color="red")
#Showing
plt.show() 
```
