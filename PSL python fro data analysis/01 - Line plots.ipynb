# Types of plots
Matplotlib allows an easy access to most of the commonly used types of plots. Let's review some of them. But first, we need to import matplotlib.

```python
import matplotlib.pyplot as plt
```

# Line plots
Line plots are used to represents a sequence of data points in a 2 dimensional graph that will be connected by lines, making it seem like a continuous line.<br />
Line plots are called using the **"plot()"** function, with an **"x"** and a **"y"** arguments.

```python
# Generating data
import numpy as np
x = np.linspace(0, 10, 1000)
y = np.sin(x)

# Initializing figure
fig = plt.figure()

# Plotting
plt.plot(x, y)

#Showing
plt.show()
```

Look how smooth it is! As I told you, the point are connected to make it seem continuous. Thus, this type of graph is particularily susceptible to **aliasing**.<br />
We generated a x sequence with 1000 points in our example. Now, try to see what hapeens if you have less points (i.e. 5 points).

```python
# TODO :Generating data
x = np.linspace(0,10,5)
y = np.sin(x)

# TODO :Plotting data
fig = plt.figure()
plt.plot(x,y)
plt.show()
```

```python
y = [10, 22, 8, 4, 15]
fig = plt.figure()
plt.plot(y)
plt.show()
```

## Matplotlib interfaces
We now know how to plot data, so it is a good time to learn about the **two types of interfaces** that matplotlib offers.<br />
The first one is a **MATLAB style interface**, using only the **"plt"** to call every plotting fonction.
As a matter of fact, it is the one that we were using above.

```python
x = np.linspace(1, 10, 1000)
fig = plt.figure()
plt.plot(x, np.exp(x))
plt.show()
```

The second type of interface is a more **object oriented style**, closer to the philosphy of Python. <br />
With this, we initialize the figure as usual.<br />
Then, we initialize **axes**, which will correspond to every graph in our figure.<br />
After, we can use any plotting function in as a **built-in function of our ax object**.

```python
x = np.linspace(1, 10, 1000)

fig = plt.figure()

# Initializing the ax of the figure
ax = plt.axes()
ax.plot(x, np.exp(x))
plt.show()
```

The latter will be particularilly useful for complicated figures construction that we will se later (i.e. with subplots,...)

## Customization
The most creative part of making figures is to play with all the parameters that allow to customize the graph.
Let's see some of them.
### Color
You can change the color of the line using the **"color"** argument, then specifying the wanted color as a string.
You can learn more about matplotlib colors [here](https://matplotlib.org/3.1.0/gallery/color/named_colors.html)

```python
x = np.linspace(1, 10, 1000)
y = np.cos(x)
fig = plt.figure()

plt.plot(x,y, color="black")

plt.show()
```

The default color rotation can be called as a string composed of an upper-case C and the index of the desired color, as follows.

```python
# Plotting in the second color of the rotation
x = np.linspace(1, 10, 1000)
y = np.sin(x)
fig = plt.figure()
plt.plot(x, y, color="C1")
plt.show()
```

### Alpha
Transparency can be added to the line using the **"alpha"** argument, then specifying the wanted transparency as a float between 0 and 1.
The default is 1 (opaque).

```python
x = np.linspace(1, 10, 1000)
y = np.cos(x)
fig = plt.figure()

plt.plot(x, y, alpha=0.5)

plt.show()
```

### Thickness
You can change the thickness using the **"linewidth"** argument, then specifying the thickness as a float.
The default thickness is 1.

```python
x = np.linspace(1, 10, 1000)
y = np.sin(x)
fig = plt.figure()

plt.plot(x, y, linewidth=10)

plt.show()
```

### Line style
You can change the style of the line using the **"linestyle"** argument, then specifying the style as a string.
The default thickness is "-".

```python
x = np.linspace(1, 10, 1000)
y = np.sin(x)
fig = plt.figure()

plt.plot(x, y, linestyle="--")

plt.show()
```

### Marker
If you want your data point to appear on the graph, you can use the **"marker"** argument, then specifying the style as a string. The size of the marker can be set using the **"markersize"** argument, as a float.

```python
x = np.linspace(1, 10, 5)
y = np.sin(x)
fig = plt.figure()

plt.plot(x, y, marker="o")

plt.show()
```

f(x)   = \frac{\sin(x)}{x}
\end{equation*}

\begin{equation*}
f(x)   = \frac{\sin(\pi x)}{\pi x}
\end{equation*}

\begin{equation*}

```python

x = np.linspace(-15, 15, 1001)
x+= 1e-10
y1 = np.sin(x)/x
y2=np.sin(np.pi*x)/(np.pi*x)
y3=np.sin(10*x)/(-5*x)

fig = plt.figure()

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.show()
```
