# Scatter plots
Scatter plots are used to represents data points in a 2 dimensional graph, where every data point will be represented individually.<br />
Scatter plots are called using the **"scatter()"** function, with an **"x"** and a **"y"** arguments.

```python
# Generating data
import numpy as np

# Initializing a random number generator (RNG)
rng = np.random.RandomState(0)

x = rng.rand(30)
y = rng.rand(30)

# Initializing figure
fig = plt.figure()

# Plotting
plt.scatter(x, y)

#Showing
plt.show()
```

## Scatter plots using the standard plot function
It is possible to plot a scatter plot using the regular plot function. For this, you just need to specifiy the **"x"** argument, the **"y"** argument, and the **marker** argument as a string.

```python
# Initializing a random number generator (RNG)
rng = np.random.RandomState(0)

x = rng.rand(30)
y = rng.rand(30)

fig = plt.figure()
plt.plot(x, y, "o")
plt.show()
```

## Taking advantage of individual data points
As mentionned above, **all data points are individual**. It means that parameters such as color, size and markers can be tweaked for each individual data points.
For this, instead of specifying one value per parameter, you need to specify an **iterable** that has the correct number of values (**the number of data points**), and compatible values.

```python
rng = np.random.RandomState(1)
x = rng.rand(100)
y = rng.rand(100)
sizes = 1000 * rng.rand(100)

fig = plt.figure()

plt.scatter(x, y, s=sizes, alpha=0.3)

plt.show()
```

It is also possible to use a **colormap** to set colors, for this you can use the color as a float. By default it will use a colormap called **"viridis"**.<br />
The colormap can be adjusted using the **cmap** argument.

```python
rng = np.random.RandomState(1)
x = rng.rand(100)
y = rng.rand(100)
#TODO: set colors and sized using random values
colors = rng.rand(100)
sizes = 1000*rng.rand(100)

fig = plt.figure()

plt.scatter(x, y, s=sizes, c=colors, alpha=0.5)

plt.show()
```

## Exercise: Hypnotyzing spiral
Lets consider a circle of center (0, 0) and radius 1.<br />
The x position of a point contained in this can be obtained using \begin{equation*}
x   = cos(theta)
\end{equation*}
where theta is the angle formed by the horizontal axis and the segment going from the center of the circle to the point.<br />
Similarily, the y position of this point can be obtained using \begin{equation*}
y   = sin(theta)
\end{equation*}

Using a list of angles theta and the scatter function, plot all the points of a circle such as 
\begin{equation*}
radius   = theta^{2}
\end{equation*}

```python
theta = np.linspace(0, 10*np.pi, 500)
r=theta**2
x=r*np.cos(theta)
y=r*np.sin(theta)



fig = plt.figure()

plt.scatter(x, y)

plt.show()
```
