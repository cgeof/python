# Histograms and Density
## 1D Histogram
Histograms are used to **visualize distributions**, by counting the **number of elements per bins**.<br />
With Matplotlib, you can plot histograms by simply calling the **"hist()"** function and passing the data as parameters.<br />

```python
# Generating data
import numpy as np
x = np.random.randn(10000)

# Initializing figure
fig = plt.figure()

# Plotting
plt.hist(x)

#Showing
plt.show()
```

Notice that we only passed the dataset as parameter, **we did not specify the binning** that we wanted. When we do this, we let matplotlib determine automatically an **appropriate number of bin**.<br />
However, we can specify the number of bins that we want using the **"bins"** parameter, followed by the number of bins that is wanted (int). The first and last edges will automatically be set as respectively the min and max of the distribution.

```python
x = np.random.randn(10000)
fig = plt.figure()
# TODO: Plot the histogram, with 60 bins 
plt.hist(x,bins=60)
plt.show()
```

Looks way smoother now! More than the number of bins that you want, you can actually **pass it directly the bins that we want**.<br />
Lets imagine that we want to start at -2, finish at 2 and have 100 bins. We could do as follows:

```python
x = np.random.randn(10000)
bins = np.linspace(-2, 2, 100)

fig = plt.figure()
plt.hist(x, bins=bins)
plt.show()
```

### Density and Cumulative
Representing an histogram as a **density** is similar to plotting the **probability density**. Effectively, it takes the count per bin and divides it by the total number of elements in the data and the bin width.<br />
The histogram can be represented as a density using the **"density"** parameter, putting it to **True**.

```python
x = np.random.randn(10000)
bins = np.linspace(-2, 2, 100)

fig = plt.figure()
plt.hist(x, bins=bins, density=True)
plt.show()
```

It is possible to represent the graph as a cumulative histogram using the **"cumulative"** parameter, putting it to True.<br />
If you want the cumulative density, you will need to put both denisty and cumulative to True. Also it is better to use "step" has histtype, for better visualization.

```python
x1 = np.random.randn(10000)
x2 = np.random.randn(10000) + 1
x3 = np.random.randn(10000) - 1
bins = np.linspace(-3, 3, 100)

fig = plt.figure()
# TODO: Represent the cumulative density histogram of x1, x2, and x3 on the same plot
#       Use "step" as histtype 
plt.hist(x1,density=True, cumulative=True, histtype="step", bins=bins)
plt.hist(x2,density=True, cumulative=True, histtype="step", bins=bins)
plt.hist(x3,density=True, cumulative=True, histtype="step", bins=bins)
plt.show()
```

## 2D Histogram
Similarily as 1D histograms used to visualize univariate distributions, two dimensional histograms can be performed to visualize bivariate distributions.<br />
To perform classical 2D histograms (square shaped bins), you can use the **"hist2d()"** function, passing it your **"x values"**, **"x values"** and the **number of bins per dimensions** that you would like.

```python
# Generating data
mean = [1, 1]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T
```

```python
fig = plt.figure()
# TODO: Generate the 2D histogram of x and y, setting bins to 50
plt.hist2d(x,y,bins=50)
# Displays the colorbar using plt.colorbar()
cb = plt.colorbar()
plt.show()
```

Note that to display 3 dimensions in a 2 dimensional plot, a colormap is used. You can change the colormap using the **"cmap"** parameter.

### Hexagonal Bins
Instead of square bins, hexagonal bins can be used. For this, you can use **"hexbin()"** function, passing it your **"x values"**, **"x values"** and the **number of bins per dimensions**(gridsize) that you would like.

```python
fig = plt.figure()
plt.hexbin(x, y, gridsize=20)
cb = plt.colorbar()
plt.show()
```
